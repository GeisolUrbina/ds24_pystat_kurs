### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Plotting can be done in two ways: pyplot (State based interface) and the "object-oriented (OO) style". 
# With pyplot we can use functions like: scatter(), plt.plot() to create and manage figures and axes. 
# The object-oriented style is recomemended for more advancerad or complex visualitations because,
# it provides clearer control over figure and axes, especially in complex visualitations, it also make the code easier to make. 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure: is the very basis for plotting data in matplotlib. When we plot data on a graph, that graph will be drawn on a figure.
# Axes: define the part of the figure where data points, lines, etc are placed.
# Artist: is any object that can be rendered on the figure. For exampel: lines, text, images, etc. 
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4)
y = x**2

x 
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)

ax.set(title="$y=x^2$", xlabel = "x", ylabel= "y")
```

```python
x = np.linspace(-4,4)
y = x**2
plt.plot(x, y, label = "y = x^2")

plt.title("y = x^2")
plt.xlabel("X-axel")
plt.ylabel("Y-axel")
plt.legend()

plt.show 
```
